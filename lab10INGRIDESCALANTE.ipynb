```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Aprendizaje supervisado
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = load_iris()
x = data.data[:, :2]
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Modelo de regresion
model = LogisticRegression()

# Entrenar datos
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# Calcular la precisiÃ³n del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precision del modelo: {accuracy: .2f}")

df = pd.DataFrame({
    "Longitud del sepalo": x_test[:, 0],
    "Ancho del sepalo": x_test[:, 1],
    "Especie real": y_test,
    "Especie predicha": y_pred
})

# Definir tamaÃ±o del grÃ¡fico
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Longitud del sepalo", y="Ancho del sepalo", hue="Especie real", data=df, palette="dark", marker="X", s=100)
plt.title("Clasificacion - Especies reales vs predichas")
plt.xlabel("Longitud del sepalo")
plt.ylabel("Ancho del sepalo")
plt.legend(title="Leyenda")
plt.show()
```
