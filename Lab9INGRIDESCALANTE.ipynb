```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

df = pd.DataFrame(data)
media = np.mean(df["Ventas"])
desvStd = np.std(df["Ventas"])
df["es_mayor_a_la_media"] = df["Ventas"] > media
plt.bar(df["Mes"], df["Ventas"], color = "orange")
plt.title("Ventas Mensuales")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.xticks(rotation = 90)
plt.axhline(media, color = "red", linestyle = "--", label = "Media de ventas")

```

```python
newData = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matematicas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

df2 = pd.DataFrame(newData)
mediaMatematicas = np.mean(df2["Matematicas"])
desvStdMatematicas = np.std(df2["Matematicas"])

mediaCiencias = np.mean(df2["Ciencias"])
desvStdCiencias = np.std(df2["Ciencias"])

mediaHistoria = np.mean(df2["Historia"])
desvStdHistoria = np.std(df2["Historia"])

df2["promedio"] = df2[["Matematicas", "Ciencias", "Historia"]].mean(axis = 1)
df2["promedio_mayor_a_90"] = df2["promedio"] > 90

plt.bar(df2["Estudiante"], df2["promedio"], color = "orange")
plt.title("Promedio de Estudiantes")
plt.xlabel("Estudiantes")
plt.ylabel("Promedio")
plt.xticks(rotation = 90)
plt.show()
```
